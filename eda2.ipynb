```python
# Replace missing values with the mean for numerical columns
numerical_columns = ['End_Lat', 'End_Lng', 'Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 
                      'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)']

for column in numerical_columns:
    mean_value = df[column].mean()
    df[column].fillna(mean_value, inplace=True)

# For categorical columns like 'Sunrise_Sunset', 'Civil_Twilight', etc., you can use mode
categorical_columns = ['Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight']

for column in categorical_columns:
    mode_value = df[column].mode()[0]  # Using [0] to get the first mode in case there are multiple modes
    df[column].fillna(mode_value, inplace=True)
```

```python
# Replace missing values with the mean for numerical columns
numerical_columns = ['Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 
                      'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)']

for column in numerical_columns:
    mean_value = df[column].mean()
    df[column].fillna(mean_value, inplace=True)

# Replace missing values with the mode for categorical columns
categorical_columns = ['Description', 'Street', 'City', 'Zipcode', 'Timezone', 'Airport_Code', 'Weather_Timestamp', 'Wind_Direction', 'Weather_Condition']

for column in categorical_columns:
    mode_value = df[column].mode()[0]  # Using [0] to get the first mode in case there are multiple modes
    df[column].fillna(mode_value, inplace=True)
```

```python
import matplotlib.pyplot as plt
severity_distribution = df['Severity'].value_counts()

plt.figure(figsize=(15, 15))
plt.pie(severity_distribution, labels=severity_distribution.index,autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Severity Levels')
plt.show()
```

```python
# Convert 'Start_Time' to datetime format
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce', infer_datetime_format=True)
```

```python
# Analyze the distribution of incidents over time (hourly)
hourly_distribution = df['Start_Time'].dt.hour.value_counts().sort_index()
```

```python
# Plot hourly distribution
plt.figure(figsize=(12, 6))
hourly_distribution.plot(kind='bar', color='skyblue')
plt.title('Distribution of Incidents by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Count')
plt.show()
```

```python
# Analyze the distribution of incidents over time (daily)
daily_distribution = df['Start_Time'].dt.dayofweek.value_counts().sort_index()
```

```python
# Plot daily distribution
plt.figure(figsize=(12, 6))
daily_distribution.plot(kind='bar', color='skyblue')
plt.title('Distribution of Incidents by Day of the Week')
plt.xlabel('Day of the Week (0 = Monday, 6 = Sunday)')
plt.ylabel('Count')
plt.show()
```

```python

# Analyze the distribution of incidents over time (monthly)
monthly_distribution = df['Start_Time'].dt.month.value_counts().sort_index()
```

```python

# Plot monthly distribution
plt.figure(figsize=(12, 6))
monthly_distribution.plot(kind='bar', color='skyblue')
plt.title('Distribution of Incidents by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
```

```python
# List of categorical variables to explore
categorical_variables = ['Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal', 'Turning_Loop']

# Plot the distribution of each categorical variable
plt.figure(figsize=(16, 8))
for i, variable in enumerate(categorical_variables, 1):
    plt.subplot(2, 3, i)
    sns.countplot(x=variable, data=df, palette='viridis')
    plt.title(f'Distribution of {variable}')
    plt.xlabel(variable)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Select numerical columns for correlation analysis
numerical_columns = ['Severity', 'Start_Lat', 'Start_Lng', 'Distance(mi)', 'Temperature(F)', 'Wind_Chill(F)',
                     'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)']

# Create a correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python


# Explore relationships between severity and other numerical variables
plt.figure(figsize=(19, 8))
for i, column in enumerate(numerical_columns[1:], 1):
    plt.subplot(3, 4, i)
    sns.boxplot(x='Severity', y=column, data=df, palette='viridis')
    plt.title(f'Relationship between Severity and {column}')
    plt.xlabel('Severity')
    plt.ylabel(column)

plt.tight_layout()
plt.show()
```

```python
# Explore relationships using box plots
plt.figure(figsize=(16, 8))
sns.boxplot(x='Severity', y='Distance(mi)', data=df, palette='viridis')
plt.title('Relationship between Severity and Distance')
plt.xlabel('Severity')
plt.ylabel('Distance (mi)')
plt.show()
```

```python

# Explore relationships using scatter plots
plt.figure(figsize=(16, 8))
sns.scatterplot(x='Temperature(F)', y='Humidity(%)', hue='Severity', data=df, palette='viridis', alpha=0.5)
plt.title('Relationship between Temperature, Humidity, and Severity')
plt.xlabel('Temperature (F)')
plt.ylabel('Humidity (%)')
plt.show()
```

```python
# Select twilight-related columns
twilight_columns = ['Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight']

# Plot the distribution of incidents for each twilight level
plt.figure(figsize=(16, 8))
for i, column in enumerate(twilight_columns, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=column, data=df, palette='viridis')
    plt.title(f'Distribution of Incidents based on {column}')
    plt.xlabel(column)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Convert 'Start_Time' to datetime format
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

# Extract the month from 'Start_Time'
df['Month'] = df['Start_Time'].dt.month

# Map months to seasons
seasons = {1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring', 6: 'Summer',
           7: 'Summer', 8: 'Summer', 9: 'Fall', 10: 'Fall', 11: 'Fall', 12: 'Winter'}
df['Season'] = df['Month'].map(seasons)

# Plot the seasonal trends
plt.figure(figsize=(12, 6))
sns.countplot(x='Season', data=df, palette='viridis')
plt.title('Seasonal Trends in Incidents')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```
